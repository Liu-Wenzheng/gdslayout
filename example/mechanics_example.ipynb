## Optomechanical_mass

### Example 1
- Device: Optomechanical_mass

Check the device

```python
from gdslayout.structure import Structure

structure = Structure(config_path="./config_example/9.yaml")
component, P = structure._load_device()
component.plot()
```

#### Show the device's ports

```python
from gdslayout.visualization import plot_component
print(component.pprint_ports())
fig, ax = plot_component(component, with_ports=True, with_labels=True, figsize=(10, 8))
```
