## Photonic crystal & waveguide coupling

### Example 1
- Device: PhC_optomechanics1
- Coupler: point

```python
from gdslayout.structure import Structure
structure = Structure(config_path="./config_example/8.yaml")
component = structure.build()
k = component.plot()
```

Check the device

```python
device, device_path = structure._load_device()
device.plot()
```

#### Show the device's ports

```python
from gdslayout.visualization import plot_component
device.pprint_ports()
fig, ax = plot_component(device, with_ports=True, with_labels=True, figsize=(10, 8))
```

#### Show the component's ports

```python
from gdslayout.visualization import plot_component
component.pprint_ports()
fig, ax = plot_component(component, with_ports=True, with_labels=True, figsize=(10, 8))
```

#### Lumerical FDTD simulation
For lumerical FDTD simulation, we need to define the coupling region and ports info.

```python
ax = structure.lumerical(plot=True, auto_compute='phc')
```

### Example 2
- Device: PhC_optomechanics2
- Coupler: port_coupler

```python
structure = Structure(config_path="./config_example/11.yaml")
component = structure.build()
component.plot()
```

Check the device

```python
device, device_path = structure._load_device()
device.plot()
```

#### Show the device's ports

```python
from gdslayout.visualization import plot_component
device.pprint_ports()
fig, ax = plot_component(device, with_ports=True, with_labels=True, figsize=(10, 8))
```

#### Show the component's ports

```python
from gdslayout.visualization import plot_component
component.pprint_ports()
fig, ax = plot_component(component, with_ports=True, with_labels=True, figsize=(10, 8))
```

#### Lumerical FDTD simulation
For lumerical FDTD simulation, we need to define the coupling region and ports info.

```python
ax = structure.lumerical(plot=True, auto_compute='phc')
ax.scatter([structure.device.ports['exciton'].x], [structure.device.ports['exciton'].y], color='red', s=10, label='Device Points')
l = structure.device.ports['exciton'].info.get('exciton_width', 0)
w = structure.device.ports['exciton'].info.get('exciton_height', 0)
ax.scatter([structure.device.ports['exciton'].x + l/2], [structure.device.ports['exciton'].y + w/2], color='red', s=10, label='Device Points')
```
