## Archimedean_super_long_spiral & tunable MZI waveguide coupling

### Example 14
- Device: spiral_archimedean1
- Coupler: dual_out_point

```python
from gdslayout.structure import Structure

structure = Structure(config_path="./config_example/14.yaml")
component = structure.build()
component.plot()
```

```python
device, device_path = structure._load_device()
device.plot()
coupler, coupler_path = structure._load_coupler()
coupler.plot()
```

```python
import matplotlib.pyplot as plt

# Plot device path curvature
plt.figure(figsize=(15, 5))
curvature = device_path.curvature()
plt.plot(curvature[0], curvature[1], label='Device Curvature')
plt.legend()
# Plot coupler path curvature in a separate figure
plt.figure(figsize=(15, 5))
curvature = coupler_path.curvature()
plt.plot(curvature[0], curvature[1], label='Coupler Curvature')
plt.legend()
```

#### Show the component's ports

```python
from gdslayout.visualization import plot_component
component.pprint_ports()
fig, ax = plot_component(component, with_ports=True, with_labels=True, figsize=(10, 8))
```

```python
structure.create_gds('./gds/14.gds')
```
